# Aprendizaje Estadístico vs. Aprendizaje Automático

Tomado de: [Aprendizaje Estadístico vs. Aprendizaje Automático](https://rubenfcasal.github.io/aprendizaje_estadistico/aprendizaje-estad%C3%ADstico-vs.-aprendizaje-autom%C3%A1tico.html)

El término **Machine Learning** (ML; Aprendizaje Automático) se utiliza en el campo de la Inteligencia Artificial desde 1959 para referirse principalmente a algoritmos de predicción, inicialmente enfocados en el reconocimiento de patrones. Muchas de las herramientas que emplea ML provienen del campo de la **Estadística**, ya que esta, junto con las **Matemáticas**, constituye la base de los enfoques para analizar datos, por lo que no conviene perder de vista esta base formal. 

Desde la **Estadística Computacional** se introdujo el término **Statistical Learning** (Aprendizaje Estadístico) para hacer referencia a este tipo de herramientas. Este es un punto de vista más estadístico donde se considera la incertidumbre inherente al no disponer de toda la información.

Tradicionalmente, ML no se preocupa por el origen de los datos y es habitual asumir que un conjunto masivo de datos equivale a disponer de toda la información (es decir, la población completa).

> “The sheer volume of data would obviate the need of theory and even scientific method”  
> — Chris Anderson, físico y periodista, 2008

Por el contrario, en **Aprendizaje Estadístico** (AE), se intenta comprender el proceso subyacente que genera los datos y si estos son representativos de la población de interés (por ejemplo, si presentan algún tipo de sesgo). No obstante, en este contexto, ambos términos se tratarán como sinónimos.

Tanto ML como AE hacen un uso intensivo de la **programación matemática**, ya que muchos de los problemas que enfrentan se formulan como optimización de funciones bajo restricciones. A su vez, los algoritmos de ML/AE también se utilizan en el ámbito de la **optimización**.

## Machine Learning vs. Data Mining

Muchas personas utilizan indistintamente los términos **Machine Learning** (ML) y **Data Mining** (DM), aunque, aunque existen diferencias sutiles entre ellos.

- **Machine Learning** es un conjunto de algoritmos dedicados principalmente a hacer predicciones y que, en su mayoría, son automáticos, minimizando la intervención humana.
  
- **Data Mining**, en cambio, busca entender los conjuntos de datos encontrando patrones, lo cual requiere una intervención humana activa, similar a la **Inferencia Estadística** tradicional. Entre otras técnicas, utiliza métodos automáticos de ML, lo que lo aproxima al **Aprendizaje Estadístico**.

## Las dos culturas (Breiman, 2001b)

Breiman distingue dos objetivos en el análisis de datos: **información** (en el sentido de inferencia) y **predicción**, cada uno de los cuales genera una cultura distinta:

### Modelización de datos
Esta cultura es propia de la estadística tradicional y se centra en el desarrollo de modelos que representen y ajusten los datos de forma que permitan hacer inferencias sobre los procesos subyacentes que generan esos datos. Aquí, el interés está en entender y modelar la estructura interna de los datos, como identificar relaciones causales, patrones y distribuciones. El objetivo no es solo predecir, sino también explicar y comprender los fenómenos estudiados. Este enfoque es común en la investigación académica y se basa en la teoría estadística, donde los modelos tienen una sólida base matemática y buscan ser interpretables.

### Modelización algorítmica

Por otro lado, la cultura de la modelización algorítmica está más orientada hacia la predicción que hacia la interpretación. Se enfoca en la creación de algoritmos capaces de hacer predicciones precisas, sin preocuparse necesariamente por los mecanismos o procesos subyacentes que generan los datos. En este enfoque, no importa tanto si el modelo es interpretable, sino si es efectivo en predecir con precisión. Este enfoque es típico de los sectores industriales y de la ingeniería informática, donde el objetivo principal es obtener resultados precisos y eficientes. Machine Learning es el núcleo de esta cultura, en la que la precisión predictiva es clave, como se ve en competiciones de algoritmos, como el Netflix Challenge, donde el éxito se mide en términos de exactitud y eficiencia de las predicciones.

## Machine Learning vs. Estadística (Dunson, 2018)

Existen diferencias significativas entre las comunidades de **Machine Learning** y **Estadística**:

- **Comunidad de ML:** tiene sus raíces en la ingeniería, la informática y, en cierta medida, en la neurociencia, emergiendo de la inteligencia artificial. Las principales publicaciones suelen ser actas de conferencias revisadas por pares, como **Neural Information Processing Systems** (NIPS), y la investigación tiende a ser rápida, enfocada en métricas de rendimiento en predicción. En los últimos años, gran parte del enfoque ha estado en métodos de redes neuronales profundas. Además, la comunidad de ML suele promocionar activamente su trabajo, buscando monetizar algoritmos a corto plazo, lo que la orienta más hacia problemas industriales.

- **Comunidad estadística:** formada principalmente por investigadores con formación matemática, cuyo enfoque es más lento y meticuloso, con énfasis en el soporte teórico (como propiedades asintóticas). Los estadísticos son más conservadores a la hora de promocionar su trabajo y su formación se basa en áreas como cálculo multivariado, álgebra lineal, ecuaciones diferenciales y análisis real. En contraste, los investigadores de ML suelen centrarse más en el procesamiento de señales, la optimización y la complejidad computacional.

# Ejemplo de tarea: estimación de $f$ en el Aprendizaje Estadístico

Dado un conjunto de predictores (características) $X_1, X_2, X_3, \dots, X_p$, existe una variable de interés $Y$ que está relacionada con \(X\), y esta relación puede ser modelada de la siguiente manera:

\begin{equation}
Y = f(X) + \epsilon
\end{equation}

donde $f(X)$ es la función que describe la relación entre los predictores y $Y$, y $\epsilon$ es un término de error que captura la variabilidad no explicada por el modelo. El **aprendizaje estadístico** se refiere al conjunto de enfoques que permiten estimar $f$, teniendo en cuenta la naturaleza de los datos y su estructura.

## ¿Cuál es la mejor forma de estimar $f$?

De acuerdo con el [teorema "No Free Lunch"](https://ti.arc.nasa.gov/m/profile/dhw/papers/78.pdf), no existe un algoritmo de optimización universal que sea superior para todos los problemas. Para cada par de algoritmos de optimización, existen tantos problemas en los que un algoritmo es mejor que el otro, como problemas en los que el segundo algoritmo es mejor que el primero. En resumen, **todos los algoritmos se comportan de manera igualmente buena o igualmente mala en promedio frente a diferentes tipos de problemas**. Esto implica que no hay un algoritmo general que sea óptimo para todos los casos, y es necesario incorporar conocimientos específicos sobre el problema para elegir el mejor enfoque.

En otras palabras:

> "Si tu única herramienta es un martillo, tratarás cada problema como si fuera un clavo"  
> — *Abraham Maslow, 1966*.

Cada algoritmo de clasificación tiene sus propios sesgos inherentes, y ninguno es superior en términos absolutos sin hacer suposiciones sobre la tarea específica. En la práctica, es necesario comparar varios algoritmos diferentes para entrenar y seleccionar el que ofrezca el mejor rendimiento. Además, es fundamental definir una métrica adecuada para medir el rendimiento de cada algoritmo y poder compararlos de manera objetiva.

# Aprendizaje Supervisado

El **aprendizaje supervisado** es una técnica de aprendizaje automático en la cual un modelo aprende a partir de un conjunto de datos etiquetados. Es decir, el conjunto de entrenamiento contiene tanto las entradas (características o predictores) como las salidas (etiquetas o valores) esperadas. El objetivo del modelo es encontrar una función \( f \) que relacione las entradas con las salidas de manera que, al recibir una nueva entrada, pueda predecir la salida correspondiente con alta precisión.

Este enfoque puede clasificarse en dos tipos:

- **Algoritmos de regresión**: Se utilizan cuando la variable de salida es continua. El objetivo es predecir valores numéricos, como precios, temperaturas, o cualquier magnitud que pueda medirse de forma continua.
  
- **Algoritmos de clasificación**: Se utilizan cuando la variable de salida es discreta o categórica. El objetivo es predecir una clase o etiqueta, como si un correo electrónico es spam o no, o si una imagen contiene un perro o un gato.

# Aprendizaje No Supervisado

El **aprendizaje no supervisado** se diferencia del supervisado en que los datos no están etiquetados. El objetivo es encontrar patrones o estructuras ocultas en los datos sin ninguna información previa sobre qué resultados esperar. Este enfoque se utiliza en situaciones donde no se tiene una variable de salida clara y se quiere analizar los datos para identificar relaciones o agrupar observaciones similares.

Algunos de los métodos más comunes incluyen:

- **Clustering (Agrupamiento)**: Agrupar los datos en subconjuntos o "clusters" de elementos similares. Se utiliza en aplicaciones como la segmentación de clientes o la agrupación de imágenes similares.
  - **K-Means**: Divide los datos en \( K \) grupos o clústeres, donde cada punto pertenece al clúster con el centroide más cercano.
  - **DBSCAN**: Algoritmo que agrupa puntos densamente conectados y detecta ruido (outliers).
  - **Análisis de Clúster Jerárquico (HCA)**: Agrupa los datos en una estructura jerárquica, desde clústeres más pequeños hasta más grandes.

- **Detección de Anomalías**: Detectar datos inusuales o atípicos en grandes conjuntos de datos. Se utiliza en aplicaciones como la detección de fraudes o la identificación de fallas en sistemas industriales.
  - **One-Class SVM**: Un tipo de SVM que se utiliza para detectar anomalías, aprendiendo a distinguir entre datos normales y anómalos.
  - **Isolation Forest**: Un algoritmo que construye múltiples árboles de decisión y aísla los puntos de datos atípicos en las primeras divisiones del árbol.

- **Reducción de Dimensionalidad**: Reducir el número de variables en un conjunto de datos mientras se preserva la información más importante. Es útil para la visualización y la preparación de datos para otros algoritmos.
  - **PCA (Análisis de Componentes Principales)**: Un método lineal que reduce las dimensiones de los datos al encontrar las direcciones de mayor varianza.
  - **Kernel PCA**: Una extensión de PCA que permite capturar relaciones no lineales entre las variables.
  - **LLE (Locally Linear Embedding)**: Un método no lineal de reducción de dimensionalidad que preserva la estructura local de los datos.

- **Aprendizaje de Reglas de Asociación**: Descubrir relaciones interesantes entre variables en grandes bases de datos, por ejemplo, en el análisis de cestas de compra para identificar productos que suelen comprarse juntos.

- **Reducción de dimensionalidad para visualización**: Proyección de datos de alta dimensionalidad en espacios de menor dimensión para facilitar su interpretación visual.

## Aprendizaje por Refuerzo

El **aprendizaje por refuerzo** es un enfoque diferente, donde un agente aprende a tomar decisiones a través de interacciones con su entorno. El agente recibe **recompensas** o **penalizaciones** según sus acciones, lo que le permite aprender una política óptima que maximice la recompensa acumulada a largo plazo. Este enfoque es ideal para problemas secuenciales, donde las decisiones en un momento dado afectan el futuro.

El agente se enfrenta a un entorno en el que debe aprender qué acciones maximizan la recompensa sin conocer de antemano la estructura del entorno. Utiliza un proceso de ensayo y error para explorar y explotar las estrategias que le permitirán alcanzar el mejor resultado.


```bash 
.
├── Aprendizaje por Refuerzo
│   ├── Métodos Basados en Valores
│   │   ├── Q-Learning
│   │   └── SARSA (State-Action-Reward-State-Action)
│   ├── Métodos Basados en Políticas
│   │   ├── REINFORCE (Monte Carlo Policy Gradient)
│   │   └── Proximal Policy Optimization (PPO)
│   ├── Métodos Actor-Crítico
│   │   ├── A2C (Advantage Actor-Critic)
│   │   └── A3C (Asynchronous Advantage Actor-Critic)
│   ├── Aprendizaje Profundo por Refuerzo
│   │   ├── Deep Q-Network (DQN)
│   │   └── Deep Deterministic Policy Gradient (DDPG)
│   ├── Modelos Basados en Modelos
│   │   ├── Dyna-Q
│   │   └── MuZero
```



  

### Elementos del Aprendizaje Reforzado
- **Agente**: El sistema o modelo que toma decisiones.
- **Entorno**: El mundo con el que el agente interactúa.
- **Acciones**: Las decisiones que el agente puede tomar.
- **Recompensas**: La retroalimentación que recibe el agente según las acciones que toma.
- **Política**: La estrategia del agente para tomar decisiones en base a las recompensas.

### Aplicaciones del Aprendizaje Reforzado
- **Robótica**: Entrenamiento de robots para realizar tareas complejas en entornos dinámicos.
- **Sistemas de recomendación dinámicos**: Ajustar las recomendaciones en tiempo real según las interacciones del usuario.
- **Videojuegos**: Creación de agentes que aprenden a jugar y mejorar en juegos sin intervención humana.
- **Vehículos autónomos**: Optimización de las decisiones de conducción para maximizar la seguridad y eficiencia.

![Modelo de Aprendizaje Reforzado](https://example.com/imagen_reinforcement)  
*Imagen ilustrativa del modelo de aprendizaje por refuerzo*



# Segunda discusion en clase

Sobre las aplicaciones listadas reconozca o proponga para un ejemplo de cada uno de los casos (supervisado, no supervisado y reforzado) cuales son los elementos de experiencia E, tareas T y medida de desempeño P del algoritmo.
